```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
from scipy.stats import linregress
```

```python
datos = pd.read_csv('/Users/mariamorris/Documents/Universidad/Monografía/Datos/Ivslambda.txt', sep="\t", header=None)
```

```python
I = np.array(datos[0])
Lam = np.array(datos[1])
```

```python
plt.rcParams['figure.facecolor'] = 'white'
sizetitle = 30
ticksize = 20
params = {'font.family':'sans-serif','font.sans-serif':'PT Sans','legend.fontsize': 20,'legend.handlelength': 2,'axes.labelsize':20, 'xtick.labelsize': ticksize, 'ytick.labelsize': ticksize}
plt.rcParams.update(params)

plt.figure(figsize=(20,10))
plt.scatter(I,Lam)
plt.title(r"$\lambda$ vs I",fontsize=sizetitle)
plt.xlabel("I (mA)")
plt.ylabel(r"$\lambda (nm)$")
plt.savefig("/Users/mariamorris/Documents/Universidad/Monografía/Imágenes/LamVsI1.png")
```

```python
plt.figure(figsize=(20,10))
lim = 60
plt.scatter(I[I<lim],Lam[I<lim], label = "Hasta "+str(lim))
plt.title(r"$\lambda$ vs I",fontsize=sizetitle)
plt.xlabel("I (mA)")
plt.ylabel(r"$\lambda (nm)$")
plt.legend()
plt.savefig("/Users/mariamorris/Documents/Universidad/Monografía/Imágenes/LamVsI2.png")


ln=45.5
ln2=56
def pendiente (a,b,x=I,y=Lam):
    r = (x>=a) & (x<b)
    xp = x[r]
    yp = y[r]
    m = linregress(xp, yp)
    rl = m[0]*xp+m[1]
    plt.plot(xp,rl,label='{}mA - {}mA the slope is {} nm/mA'.format(a,b,round(m[0]*1000,2)))
    plt.legend()

pendiente(0,ln)
pendiente(ln,ln2)
plt.ylim([406.76,406.87])
```

```python
plt.figure(figsize=(20,10))
lim1 = 60
lim2 = 71
plt.scatter(I[(I>=lim1) & (I<lim2)],Lam[(I>=lim1) & (I<lim2)], label = str(lim1)+"<I<"+str(lim2))
plt.title(r"$\lambda$ vs I",fontsize=sizetitle)
plt.xlabel("I (mA)")
plt.ylabel(r"$\lambda (nm)$")
plt.legend()
plt.savefig("/Users/mariamorris/Documents/Universidad/Monografía/Imágenes/LamVsI3.png")
```
