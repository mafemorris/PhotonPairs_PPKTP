### Índices de refracción del cristal

```python
def nz(lam,T):
    A = 2.12725
    B = 1.18431
    C = 5.14852*10**-2*um**2
    D = 0.6603
    E = 100.00507*um**2
    F = 9.68956*10**-3*um**-2
    T0 = 25
    a1 = array([9.9587,9.9228,-8.9603,4.1010])*10**-6
    a2 = array([-1.1882,10.459,-9.8136,3.1481])*10**-8
    nr = sqrt(A+(B/(1-C/lam**2))+(D/(1-E/lam**2))-F*lam**2)
    n1 = sum([a1[i]/(lam/um)**i for i in range(len(a1))],axis=0)
    n2 = sum([a2[i]/(lam/um)**i for i in range(len(a2))],axis=0)
    dn = n1*(T-T0)+n2*(T-T0)**2
    return nr + dn
```

```python
def aja(T,lp): #Condicion de QPM debe ser = 0, en el caso degenrado 
    c = 299792458 #velocidad de la luz 
    lamd = 3.425*um #Periodo del cristal
    ths = 2 #Angulo de emision 
    ls = 2*lp #longitud de onda degenerada 
    gr = (2*pi/lp)*(nz(lp,T)-nz(ls,T)*(ths**2/2-1))-2*pi/lamd #condicion de QPM degenrada 
    return gr

T = linspace(1,130,100) #Rango de temperaturas 
lp = linspace(0.4043*um,0.408*um,20) #Rango de longitudes de onda

starting_guess = -2 #Temperatura en a que empieza a buscar 

c=[fsolve(aja, starting_guess, l)[0] for l in lp] #Temperaturas de cada longitud de bombeo 

cR = lambda l: round(fsolve(aja, starting_guess, l)[0],3) #T de QPM para el caso degenerado de l de bombeo
aCC = lambda l: plot(l*ones(len(T)),T, label = str(l) + "nm a T=" + str(cR(l*nm)) + r"$^o$C") #Plotea la linea horizontal en la t de QPM para la longitud l

figure()
plot(lp/nm,c,'k--', label="$T_{PM}$")
aCC(404.61)
aCC(404.7)
aCC(405.448)
aCC(405.458)
aCC(404.822)
aCC(404.79)
aCC(405.02)
aCC(405.16)
aCC(405.6)
xlabel(r"$\lambda_p$ [nm]")
ylabel(r"$T_{PM}$ [$^oC$]")
title(r"$T_{PM}(\lambda_p)$")
legend()
savefig(ruta+"TvsLam.png")
```

```python
def theta(T,lamp): #angulo de emision en el caso degenerado
    lamd = 3.425*um
    lamp2 = lamp/nm
    #a = np.sqrt(2*(1+(1/lamd-nz(lamp,T)/lamp2)/(nz(lamp*2,T)/lamp2)))
    a = np.sqrt(2*(1+(lamp2/(nz(lamp,T)*lamd))-(nz(lamp,T)/nz(lamp*2,T))))
    #a = np.sqrt((1/lamd)+((nz(2*lamp,T)-nz(lamp,T))/lamp2)/(nz(lamp*2,T)/(2*lamp2)))
    return a
T = np.linspace(10,25,100)
plt.plot(T,theta(T,0.4048*um))
```

```python
def L (T):#Cambio en la longitud del cristal con la temperatura 
    al = 6.7*10**-6
    be = 11*10**-9
    l0 = 15
    T0 = 25
    l= l0*(1+al*(T-T0)+be*(T-T0)**2)
    return l
plot(mm,L(mm))
y = array([20,25,50,100,150,200])
plot(y, L(y),'.r')
xlabel(r"T [$^oC$]")
ylabel("L(T) [mm]")
title("L(T)")
savefig("LvsT.png")
```
