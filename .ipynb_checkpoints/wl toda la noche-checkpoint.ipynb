```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
plt.rcParams['figure.facecolor'] = 'white'
sizetitle = 24
ticksize = 22
params = {'font.family':'sans-serif','font.sans-serif':'PT Sans','legend.fontsize': 18,'legend.handlelength': 2,'axes.labelsize':20, 'xtick.labelsize': ticksize, 'ytick.labelsize': ticksize}
plt.rcParams.update(params)
```

```python
antes = '/Users/mariamorris/Documents/Universidad/Monografía/Datos/Mi_compu_40mW/anillos/'
despues = '/Users/mariamorris/Documents/Universidad/Monografía/Imágenes/Resultados/'
data = pd.read_csv(antes+'all night 15-16 junio.txt',delimiter = "\t",
                   encoding= 'unicode_escape')
data['Power']=data['Power'].apply(lambda x: float(x.replace(',','.')))
data['Wavelength']=data['Wavelength'].apply(lambda x: float(x.replace(',','.')))
data['Time']=data['Time']/(3.6*10**6)
```

```python
plt.figure(figsize=(20,8))
upx=7
downx=2
downy = min(data['Wavelength'])-0.1
upy = max(data['Wavelength'])
plt.plot(data['Time'],data['Wavelength'],'k')
wl1 = 30000
wl2 = 50000
wl3 = 70000
for i in [30000,50000,60000,70000,80000]:
    plt.plot(data['Time'],np.ones(len(data['Wavelength']))*data['Wavelength'][i],label='{}nm at {}h'.format(round(data['Wavelength'][i],3),round(data['Time'][i],1)))
plt.ylabel(r'$\lambda_p$(nm)')
plt.xlabel(r'time (hrs)')
plt.title(r'Pump wavelength vs time (5hrs)',fontsize=sizetitle)
plt.legend()
plt.xlim([downx,upx])
plt.ylim([downy,upy])
plt.savefig(despues+'PvsT_large.png')
```
